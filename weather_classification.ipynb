```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv("/content/weather_classification_data.csv")
weather.head()
```

```python
weather_categorical = weather.loc[:,["Cloud Cover", "Season", "Location", "Weather Type"]]
weather_categorical.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoding = pd.DataFrame(weather_categorical["Weather Type"])
label_encoding['Weather Type_label'] = LabelEncoder().fit_transform(label_encoding['Weather Type'])
```

```python
One_Hot_encoding =pd.DataFrame(weather_categorical["Season"])
one_hot = pd.get_dummies(One_Hot_encoding['Season'])

df = pd.concat([One_Hot_encoding, one_hot], axis= 1)
```
